### Part 1

Write a function which applies a Bonferroni correction and which returns a Boolean vector where ```True``` indicates that the result is significant after correction, and ```False``` indicates that the statistical test is not significant. How many significant results remain in your sample?

In the _Bonferroni_ correction we only accept those tests as rejecting the null model where the $p$-value is less than $0.05/n$, where $0.05$ is the conventional threshold for a single test (other thresholds may be set), and $n$ is the total number of tests. 

````julia
# [Write your code in this cell]
# 
pBonferroni = function(pvalvec)
    pvalsS = []
    alpha = 0.05
    n = 100
    newCutoff = alpha/n
    for i in 1:size(pvalvec,1)
        if pvalvec[i] < newCutoff
            push!(pvalsS, true)
        else
            push!(pvalsS, false)
        end
    end
    return pvalsS
end
    
bpvals = pBonferroni(pvals);
number_of_significant_results = sum(bpvals)

# check:
# count(x->x<0.05, pvals) # = 42
# count(x->x<0.005, pvals) # = 23 -- Is 0.005 the alpha-value used above???
````

### Part 2

Write a function which applies a Benjamini-Hochberg correction and which returns a Boolean vector where ```True``` indicates that the result is significant after correction, and ```False``` indicates that the statistical test is not significant. How many significant results remain in your sample?

The Bonferroni correction is often found to be unnecessarily conservative; the _Benjamini-Hochberg_ correction aims to control the _False Discovery Rate_ (FDR) at a specific level, $q$. To do so we order the $p$-values such that $p_1\le p_2\le \ldots \le p_n$. 

For the desired FDR, $q$, we then find the value of $k$ such that $p_k\le \frac{k}{n}q$. For these tests we reject the null hypothesis.


````julia
# [Write your code in this cell]
# 
pBenjHoch = function(pvalvec,q=0.05)
    pvalsS2 = []
    pvalvec_sorted = sort(pvalvec)
    k = 1
    n = 100
    while (k <= size(pvalvec,1))
        if (pvalvec_sorted[k] <= k/n * q)
            push!(pvalsS2, true)
        else
            push!(pvalsS2, false)
        end
        k += 1
    end   
    return pvalsS2
end

bhpvals = pBenjHoch(pvals);
number_of_significant_results = sum(bhpvals)
````

## Problem 3:  A Model of Gene Expression 

We consider the case where mRNA, $m(t)$, changes according to the simple differential equation,

$$
  \frac{dm(t)}{dt} = \alpha +f(t) -\beta m(t),
$$

and the corresponding protein abundance, $p(t)$, follows

$$
    \frac{dp(t)}{dt}=\gamma m(t) - \delta p(t).
    $$

The function $f(t)$ described the behaviour of a transcription factor. $\alpha$ is the basal expression rate, $\gamma$ the rate of translation into protein, and $\beta$ and $\delta$ are the degradation rates for mRNA and protein, respectively.

````julia
# [Write your code in this cell]
# 
function mRNAproteinODEs(t,u,du, alpha=0.1, beta=1., gamma=0.3, delta=0.1)
    du[1] = alpha + f(t) - beta*u[1]
    du[2] = gamma*u[1] - delta*u[2]
end
````

### Part1: 

Write a Gillespie algorithm for this reaction (not a general purpose solver).

````julia
# [Write your code in this cell]
# 
function gillespie(k=[5 1 10], x=30, y=3)
    t = 0;
    S = [1 -1 -1; 0 0 1] # stoichiometry matrix S
    ts = Float64[]
    xs = []
    ys = []
    k1 = k[1]
    k2 = k[2]
    k3 = k[3]
    while (t<100)
        # 1. Generates a random number r1 picked from a uniform distribution on (0,1) to determine which reaction occurs next
        r1 = rand(Uniform(0,1))
        # 2. a. Computes propensity/hazard function for each node (here: x and y) and reaction hi
        h1 = k1
        h2 = k2*x
        h3 = k3*x*y
        #  2. b. Computes total propensity/hazard of system
        h0 = h1 + h2 + h3
        # 3. Computes a second random number tau from an exponential distribution (w/ mean 1/h0) to determine time until next reaction takes place; 
        #    updates t
        tau = rand(Exponential(1/h0))
        t = t+tau
        # 4. Computes which reaction occurs at time t+tau
        if (h1/h0) > r1
            i = 1
        elseif ((h1/h0) <= r1) && (((h1+h2)/h0) > r1)
            i = 2
        else
            i = 3
        end
        # 5. Updates node states and propensities for the i-th reaction of node x and node y
        x += S[1,i] 
        y += S[2,i] 
        push!(ts, t)
        push!(xs, x)
        push!(ys, y)
        # 6. ...and now repeats it all from step 1 
    end
    return (ts, xs, ys)
end

# ts, xs, ys = gillespie()
# print("# time steps: ", length(ts), "\n")
# print("time steps: ", ts)

# plot(ts,[xs, ys], label = ["x" "y"])
````

### Part2: 
Solve the system for $[k_1,k_2,k_3]=[5,1,10]$ for $0\le t\le 100$ and intitial conditions $[x,y]=[30,3]$: simulate 1000 repeat runs and plot the mean and variance for $x$ and $y$.


````julia
# [Write your code in this cell]
# 

# Assume we impose the equidistant time grid from above (linspace(5,100,20)) and plot the mean/variance of x and y 
# over 1000 gillespie simulations at *these* given time points.
# We use no binning but only take the values at the (random) time point before we cross the given time grid value;
# the amount of particles at this random time point equals the amount of particles at the imposed time point since 
# simulation is discrete. 

# Note: We do not manipulate the gillespie function (see above) itself to only return the relevant values, but choose to 
# return all values in the simulations and then store the relevant ones in two big matrices. This seems more elegant or 
# user-friendly but might computationally be more expensive.
# 

time=collect(linspace(5,100,20))

mastermatrixXs = zeros(1000, 20)
mastermatrixYs = zeros(1000, 20)
r = 1
# simulate 1000 gillespie runs
while (r<=1000)
    ts, xs, ys = gillespie()
    t = 5 #  emulates our given grid linspace(5, 100, 20) to avoid another for-loop
    s = 1
    while (s <= length(ts)-1) # we only go through list once (since time steps monotonous, ie we don't go back in time)
        if ts[s] <= t && ts[s+1] > t # if we are close enough to our given time step, ie we skip over it in the next iteration
            mastermatrixXs[r, Int(t/5)] = xs[s] # ..we store the x and y values at this point
            mastermatrixYs[r, Int(t/5)] = ys[s]
            t += 5 
        end
        s += 1
    end
    r += 1
end

# list comprehension! yayh \o/
muX = [mean(mastermatrixXs[:,i]) for i in 1:size(mastermatrixXs,2)]
muY = [mean(mastermatrixYs[:,i]) for i in 1:size(mastermatrixYs,2)]
varX = [var(mastermatrixXs[:,i]) for i in 1:size(mastermatrixXs,2)]
varY = [var(mastermatrixYs[:,i]) for i in 1:size(mastermatrixYs,2)];

# plot
p1 = plot(time, muX, label = "", xaxis = ("time"), yaxis = ("mean of X", (0, .02)), title = "Mean and variance for X and Y")
p2 = plot(time, muY, label = "", xaxis = ("time"), yaxis = ("mean of Y"))
p3 = plot(time, varX, label = "", xaxis = ("time"), yaxis = ("variance of X", (0, .02)))
p4 = plot(time, varY, label = "", xaxis = ("time"), yaxis = ("variance of Y"))
plot(p1, p2, p3, p4, layout=(2,2), legend=false)
````
