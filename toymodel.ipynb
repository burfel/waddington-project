```julia
#using DataFrames, Plots, StatPlots, Distributions, DifferentialEquations

using Distributions, Plots, DifferentialEquations

a1=1
a2=1
b1=1
b2=1
k1=1
k2=1
n=4
S=0.5
```

```julia

function lorenz(t,u,du)
    du[1] = a1*u[1]^n/(u[1]^n+S^n)+b1*S^n/(u[2]^n+S^n)-k1*u[1]
    du[2] = a2*u[2]^n/(u[2]^n+S^n)+b2*S^n/(u[1]^n+S^n)-k2*u[2]
end

function σ_lorenz(t,u,du)
#    du[1] = 2
#    du[2] = 2
    du[1] = sqrt(abs(a1*u[1]^n/(u[1]^n+S^n)+b1*S^n/(u[2]^n+S^n)-k1*u[1]))
    du[2] = sqrt(abs(a2*u[2]^n/(u[2]^n+S^n)+b2*S^n/(u[1]^n+S^n)-k2*u[2]))
end
```

```julia
h=20
Ps=zeros(eye(h))
xspan=linspace(0,3,h)

total=0

for i = 0:100000
    
    x1init=rand([0.0,3.0])
    x2init=rand([0.0,3.0])
    prob_sde_lorenz = SDEProblem(lorenz,σ_lorenz,[x1init,x2init],(0.0,25.0))
    sol = solve(prob_sde_lorenz)

    for t = 1:size(sol.u,1)
        x1=1
        x2=1
        found1=false
        found2=false
        impossible=false

        while(!(found1 && found2))
            if (sol.u[t][1]<xspan[1] || sol.u[t][1]>=xspan[h] || sol.u[t][2]<xspan[1] || sol.u[t][2]>=xspan[h])
                impossible=true
                break;
            end
            
                
            if !(xspan[x1]<=sol.u[t][1] && sol.u[t][1]<xspan[x1+1]) && !found1
                x1=x1+1
            else
                found1=true
            end
                
            if !(xspan[x2]<=sol.u[t][2] && sol.u[t][2]<xspan[x2+1]) && !found2
                x2=x2+1
            else
                found2=true
            end
        end
        
        if !impossible
            Ps[x1,x2]=Ps[x1,x2]+1
            total=total+1
        end
    end
end

Ps=Ps/(total*(4/h)^2)+0.00001

Ps=-log.(Ps)

#surface(xspan,xspan,Ps)           
#plot(sol,vars=(1,2))
#end
```
